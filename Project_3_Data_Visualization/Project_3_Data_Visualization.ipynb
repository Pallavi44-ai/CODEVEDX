# **Project 3 - Data Visualization with Python**

# Superstore Sales Dataset

 Objective:

To visualize sales and profit data using different chart types and communicate business insights effectively.

Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (10,5)
```

Load Dataset

```python
df = pd.read_csv("Superstore.csv", encoding="latin1")
```

Chart 1: Sales by Category (Bar Chart)

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

Chart 2: Profit by Region (Bar Chart)

```python
region_profit = df.groupby("Region")["Profit"].sum()

region_profit.plot(kind="bar")
plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

Chart 3: Monthly Sales Trend (Line Chart)

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.month

monthly_sales = df.groupby("Month")["Sales"].sum()

monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

Chart 4: Sales Distribution (Histogram)

```python
sns.histplot(df["Sales"], bins=30)

plt.title("Sales Distribution")
plt.show()
```

Chart 5: Sales vs Profit (Scatter Plot)

```python
sns.scatterplot(x="Sales", y="Profit", data=df)

plt.title("Sales vs Profit")
plt.show()
```

Chart 6: Category Share (Pie Chart)

```python
category_sales.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Sales Contribution by Category")
plt.ylabel("")
plt.show()
```

# Key Insights

1. Technology generated the highest sales.

2. The West region produced the highest profit.

3. Monthly sales showed noticeable fluctuations.

4. Most transactions occurred in lower sales ranges.

5. Sales and profit generally showed a positive relationship.

6. Technology contributed the largest share of total revenue.

# Conclusion

Data visualization helped transform raw sales data into meaningful visual insights. Various chart types were used to understand category performance, regional profitability, sales trends, and customer purchasing behavior.
